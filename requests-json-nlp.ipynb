3. Write a function that takes an artist, song, and filename, accesses the lyrics.ovh api to get the song lyrics, and writes the results to the specified filename.  Test this function by getting the lyrics to any four songs of your choice and storing them in different files.

```python
#create function to get lyrics and put into json

def get_lyrics(artist_name, song_title, filename):
    artist = genius.search_artist(artist_name, max_songs=3, sort="title")
    song = artist.song(song_title)
    lyrics = song.lyrics
    song_dict = {'artist': artist_name, 'song': song_title, 'lyrics': lyrics}
    #write dict to a json file
    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(song_dict, f, ensure_ascii=False, indent=4)
```

```python
#get lyrics for 4 songs
get_lyrics('Taylor Swift', 'Cruel Summer', 'cruel.json')
get_lyrics('Taylor Swift', 'Paper Rings', 'paperrings.json')
get_lyrics('Taylor Swift', 'Cardigan', 'cardigan.json')
get_lyrics('Taylor Swift', 'Vigilante Shit', 'vigilante.json')
```

4. Write a function that takes the name of a file that contains song lyrics, loads the file, performs sentiment analysis, and returns the polarity score.  Use this function to print the polarity scores (with the name of the song) of the three files you created in question 3.  Does the reported polarity match your understanding of the song's lyrics? Why or why not do you think that might be?  Answer the questions in either a comment in the code cell or a markdown cell under the code cell.

```python
def sentiment(filename):
    #open json file
    f = open(filename)
    song_dict = json.load(f)
    f.close()
    
    #use spacy for sentiment analysis of lyrics
    nlp = spacy.load('en_core_web_sm')
    nlp.add_pipe('spacytextblob')
    text = song_dict['lyrics']
    doc = nlp(text)
    return doc._.blob.polarity
```

```python
sentiment('cruel.json')
```

```python
print(f"The polarity for Cruel Summer by Taylor Swift is {sentiment('cruel.json')}.") 
```

```python
print(f"The polarity for Paper Rings by Taylor Swift is {sentiment('paperrings.json')}.") 
```

```python
print(f"The polarity for Cardigan by Taylor Swift is {sentiment('cardigan.json')}.") 
```
